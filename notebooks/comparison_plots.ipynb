```python
import matplotlib.pyplot as plt
```

```python
eng_labels = ["BERT", "word2vec + LSTM", "word2vec + SVM", "word2vec + drzewo decyzyjne"]
eng_results = [0.987, 0.969, 0.978, 0.856]

fig = plt.figure(figsize=(7, 5))
fig.tight_layout(pad=10)

ylim = 1.1
val_shift = 0.02

plt.bar([i+1 for i in range(len(eng_labels))], eng_results, tick_label=eng_labels)
plt.xticks(rotation = 45)
plt.ylabel('Dokładność')
plt.ylim(0, ylim)

for i in range(len(eng_labels)):
    plt.text(i+0.85, eng_results[i]+val_shift, eng_results[i])
```

```python
pl_labels = ["word2vec + LSTM", "word2vec + SVM", "word2vec + drzewo decyzyjne", \
              "TF-IDF + SVM", "TF-IDF + drzewo decyzyjne"]
pl_results_5c = [0.54, 0.465, 0.341, 0.552, 0.396]

fig = plt.figure(figsize=(10, 5))
fig.tight_layout(pad=10)

plt.bar([i+1 for i in range(len(pl_labels))], pl_results_5c, tick_label=pl_labels)
plt.xticks(rotation = 45)
plt.ylabel('Dokładność')
plt.ylim(0, ylim)

for i in range(len(pl_labels)):
    plt.text(i+0.85, pl_results_5c[i]+val_shift, pl_results_5c[i])
```
